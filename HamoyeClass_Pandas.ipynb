CHANGING INDEXING
Pandas allows for renaming index of list other than the default 0 1 2 3 etc.

```python
scores = [400, 200, 600, 800]
myS = pd.Series(scores, index = ["math", "eng", "fre","sos"])
```

```python
print (myS)
```

```python
myS["math"]
```

**DATA** **FRAMES** **IN** **PANDAS**

Datframes in Pandas is like multidimensional data arrays (2d,3d) in numpy

```python
scores = [400, 200, 600, 800]
Dt = pd.DataFrame(scores)
```

```python
print(Dt)
```

To specify the heading of the column instead of the default 0

```python
Dt = pd.DataFrame((scores), columns = ["numbers"])
```

```python
print (Dt)
```
