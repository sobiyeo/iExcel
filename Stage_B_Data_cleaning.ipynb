```python
nan_replace = {'children':0.0, 'country':'Unknown', 'agent':1000, 'company': 0}
```

```python
df = df.fillna(nan_replace)
```

```python
# Checking for unique values in a column
for col in df.columns:
  if df[col].nunique()<20:
    print(f'The column {col} has the following values: {df[col].unique()} ')
```

```python
df['meal'].replace('Undefined','SC', inplace=True)
#replacing "undefined" with SC
```

```python
#Some rows contains entries with 0 adults, 0 children and 0 babies. which means no guests
# So I'm dropping them.
zero_guests = list(df.loc [df['adults']
+df['children']
+df['babies'] ==0].index)

df.drop(df.index[zero_guests])
```
