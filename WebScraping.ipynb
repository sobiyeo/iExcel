```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
```

```python
url = 'https://understat.com/league/'

leagues = ['EPL', 'La liga', 'Serie A', 'Bundesliga', 'Ligue 1']

years = ['2019','2020', '2021']
```

```python
final_url = url + leagues[0] +'/'+ years[0]
```

```python
resp = requests.get(final_url)
```

```python
print(final_url)
```

```python
resp.content
```

```python
soup =bs(resp.content, 'html.parser')
```

```python
soup.find_all('div', ('wrapper'))[0].find_all('div',('block'))
```

```python
#soup.find_all('script') - This will return all scripts
#to pick the second data
data = soup.find_all('script')[2]
```

```python
type(data)
```

```python
#converting to string
data = str(data)
```

```python
data = data[38:-13]
```

```python
enc_data = data.encode("utf8").decode("unicode_escape")
```

```python
loaded_data = json.loads(enc_data)
```

```python
loaded_data
```

```python
loaded_data.keys()
```

```python
loaded_data['71'].keys()
```

```python
teams =[]
for k in loaded_data.keys():
  print(loaded_data[k]['title'])
  teams.append(loaded_data[k]['title'])
```

```python
## Putting the data in a pandas dataframe
pd.DataFrame(loaded_data['71']['history'])
```

```python
team_d = pd.DataFrame(loaded_data['71']['history'])
pd.DataFrame(team_d.select_dtypes(exclude='object').sum()).T
#This select all data tyoes in the teams data and removes all non-numeri  data, adds all
#together and transposed it . i.e turns rows into columns and vice versa
```

```python
final_df = pd.DataFrame()
for k in loaded_data.keys():
  team_data = pd.DataFrame(loaded_data[k]['history'])
  numerical_d = pd.DataFrame(team_data.select_dtypes(exclude = 'object').sum().round(2)).T
  final_df = pd.concat([final_df, numerical_d], axis = 0)
   #this removes all the string and non-numerical data
```

```python
final_df
```

```python
final_df.index = teams
final_df
```

```python
#sorting in ascending order acording to points
final_df.sort_values(by = 'pts', ascending = False)
```
